```python
#Write a python program for the calculations of Point Estimates, Interval Estimate,Hypothesis testing (Z & T)
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
def inferential_stats(csv, sample_mean, std_dev, sample_size, population_mean, option, flag = 1):
    z_Dict = {0.05 : 1.68, 0.025 : 1.96}
    conclusion = "Accepting the null hypothesis"
    t_dict = {(29, 0.05) : 1.6991 , (29, 0.025) : 2.0452, (40, 0.05) : 1.6839 , (40,0.025) : 2.0211}
    alpha = 0.025
    if flag:
          print("CSV Based")  
          plantgrowth_df = pd.read_csv(csv)
          plant_weight = plantgrowth_df.describe()['weight']
          sns.histplot(plantgrowth_df['weight'], kde=True)
          plt.show()
          sample_size = plant_weight['count']
          std_dev = plant_weight['std']
          point_estimate = plant_weight['mean']
          degree_of_freedom = sample_size - 1
          if option == "Estimations" or option == "Both":
              #Point Estimate Calculation 
              if sample_size >= 30:
                  print("This data Follows Z-distribution")
                  lower_estimate = point_estimate - z_Dict[alpha] * (std_dev / math.sqrt(sample_size))
                  upper_estimate = point_estimate + z_Dict[alpha] * (std_dev / math.sqrt(sample_size)) 
              else :
                  print("This data Follows T-distribution")
                  lower_estimate = point_estimate - t_dict[(degree_of_freedom, alpha)] * (std_dev / math.sqrt(sample_size))
                  upper_estimate = point_estimate + t_dict[(degree_of_freedom, alpha)] * (std_dev / math.sqrt(sample_size))

              interval_estimate = (str(lower_estimate) + str("-") + str(upper_estimate))
              print(f"Point Estimate found to be {point_estimate}")
              print(f"Interval Estimate found to be {interval_estimate}")

          if option == "Hypothesis Testing" or option == "Both":
              population_mean = 4 
              print("Null Hypothesis : Mean Weight growth of plant is 4")
              print("Alternate Hypothesis : Mean Weight growth of plant is not 4")
              if sample_size >= 30:
                 print("Performing two-tail Z-Test")
                 z_actual = (point_estimate - population_mean) / (std_dev / math.sqrt(sample_size))
                 if z_actual > z_Dict[alpha]:
                    conclusion = "Rejecting the null hypothesis" 
              else:
                   print("Performing two-tail T-Test")
                   z_actual = (point_estimate - population_mean) / (std_dev / math.sqrt(sample_size))
                   if z_actual > t_dict[(degree_of_freedom , alpha)]:
                      conclusion = "Rejecting the null hypothesis" 
              print(f"z_actual {z_actual}")
              print(f"{conclusion} with 95% confidence")

    else:
          #Data Taken From Metro EMS
          print("Non-CSV Based")  
          if option == "Estimations" or option == "Both":
             #Point Estimate Calculation 
              point_estimate = sample_mean
              if sample_size >= 30:
                  print("This data Follows Z-distribution")
                  lower_estimate = point_estimate - z_Dict[alpha] * (std_dev / math.sqrt(sample_size))
                  upper_estimate = point_estimate + z_Dict[alpha] * (std_dev / math.sqrt(sample_size))
                 
              else :
                 print("This data Follows T-distribution")
                 degree_of_freedom = sample_size - 1
                 lower_estimate = point_estimate - t_dict[(degree_of_freedom, alpha)] * (std_dev / math.sqrt(sample_size))
                 upper_estimate = point_estimate + t_dict[(degree_of_freedom, alpha)] * (std_dev / math.sqrt(sample_size))
              
              interval_estimate = (str(lower_estimate) + str("-") + str(upper_estimate))

              print(f"Point Estimate found to be {point_estimate}")
              print(f"Interval Estimate found to be {interval_estimate}")

          if option == "Hypothesis Testing" or option == "Both":
              print("Null Hypothesis : EMS service goal is reaching in or less than 12 mins")
              print("Alternate Hypothesis : EMS service goal is reaching in more than 12 mins")
              if sample_size >= 30:
                 z_actual = (sample_mean - population_mean) / (std_dev / math.sqrt(sample_size))
                 if z_actual > z_Dict[alpha * 2]:
                    print("Performing one-tail Z-Test")
                    conclusion = "Rejecting the null hypothesis"    
              else:
                   print("Performing one-tail T-Test")
                   z_actual = (sample_mean - population_mean) / (std_dev / math.sqrt(sample_size))
                   if z_actual > t_dict[(degree_of_freedom , alpha * 2)]:
                      conclusion = "Rejecting the null hypothesis"
              print(f"z_actual {z_actual}")
              print(f"{conclusion} with 95% confidence")


print("What you want to do (Estimations or Hypothesis Testing or Both)??")
option = input() 
sample_mean = 13.25
std_dev = 3.2
sample_size = 40
population_mean = 12
h = input("By Default it is non-csv based , Do u want CSV based Inference calculation Y or N?")
if h == 'Y':
    inferential_stats("PlantGrowth.csv", sample_mean, std_dev, sample_size, population_mean, option)   
else: 
     inferential_stats("PlantGrowth.csv", sample_mean, std_dev, sample_size, population_mean, option, 0)        
             
```
