```python
#Write a Python program to select the rows where number of
#attempts in the examination is less than 2 and score greater than
#15.
df = pd.DataFrame({
           'Examination' :['IAS', 'NEET', 'CAT', 'JEE MAINS',],
           'Attempts' :[2,0,1,6],
           'Score' : [15,20,100,99]})

df[(df.Attempts<2) & (df.Score>15)]
```

```python
#Write a Python program to append a new row 'k' to Data Frame
#with given values for each column. Now delete the new row and
#return the original data frame.

df = pd.DataFrame({
           'Examination' :['IAS', 'NEET', 'CAT', 'JEE MAINS',],
           'Attempts' :[2,0,1,6],
           'Score' : [15,20,100,99]})

#Adding new row to dataframe
df.loc['K'] = ['GMAT',19, 90]
print(df)
#Now deleting new row and returning old dataframe
df.drop(labels='K', inplace=True)
print(df)
```

```python
#Write a Python program to sort the data frame first by 'name' in
#descending order, then by 'score' in ascending order.
import pandas as pd
df = pd.DataFrame({
            'Name' : ['Harry', 'Abbay', 'Bunty', 'Jill'],
           'Examination' :['IAS', 'NEET', 'CAT', 'JEE MAINS',],
           'Attempts' :[2,0,1,6],
           'Score' : [15,20,100,99]})

df.sort_values(by='Name',ascending=False)
```

```python
#Write a Python program to insert a new column in existing Data
#Frame.
df = pd.DataFrame({
            'Name' : ['Harry', 'Abbay', 'Bunty', 'Jill'],
           'Examination' :['IAS', 'NEET', 'CAT', 'JEE MAINS',],
           'Attempts' :[2,0,1,6],
           'Score' : [15,20,100,99]})

df['Result'] = pd.Series(['Pass', 'Failed', 'Reappear', 'Improving'])
df
```

```python
#Write a Python program to change the name 'James' to 'Suresh'
#in name column of the data frame.

df = pd.DataFrame({
            'Name' : ['James', 'Abbay', 'James', 'Jill'],
           'Examination' :['IAS', 'NEET', 'CAT', 'JEE MAINS',],
           'Attempts' :[2,0,1,6],
           'Score' : [15,20,100,99]})

df.loc[df[df.Name == 'James'].index, 'Name'] ='Suresh'
df
```
