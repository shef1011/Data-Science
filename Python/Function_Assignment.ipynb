```python
# Write a function to convert temperature from Celsius to Fahrenheit scale.
# oC to oF Conversion: Multipy by 9, then divide by 5, then add 32.
def Cel2Fah(x):
    a = (9 * x)/5 + 32
    return str('%.2f'%a)
a = float(input('Enter temperature in celsius: '))
print(Cel2Fah(a))
# Note: Return a string of 2 decimal places.
# In - Cel2Fah(28.0)
# Out - '82.40'
# In - Cel2Fah(0.00)
# Out - '32.00'
```

```python
# Write a function that uses a default value.
# In - introduce('Lim', 20)
# Out - 'My name is Lim. I am 20 years old.'
# In - introduce('Ahmad')
# Out - 'My name is Ahmad. My age is secret.'

def introduce(name, age='secret'):
    if age is not 'secret':
      return f'My name is {name}. I am {age} years old'
    else:
      return f'My name is {name}. My age is {age}'

print(introduce('Lim', 20))
print(introduce('Ahmad'))

```

```python
# For a quadratic equation in the form of ax2+bx+c, the discriminant, D is b2-4ac. Write a function to compute the discriminant, D.
# In - quadratic(1, 2, 3)
# Out - 'The discriminant is -8.'
# In - quadratic(1, 3, 2)
# Out - 'The discriminant is 1.'
# In - quadratic(1, 4, 4)
# Out - 'The discriminant is 0.'
def quadratic(a,b,c):
    d = b ** 2 - (4 * a * c)
    return d
a = int(input('Enter 1st coefficent : '))
b = int(input('Enter 2nd coefficent: '))
c = int(input('Enter constant: '))
print(f'The discriminant is {quadratic(a,b,c)}.')

```

```python
 # Write a function find_max that accepts three numbers as arguments and returns the largest number among three. 
# Write another function main, in main() function accept three numbers from user and call find_max.

def find_max(a,b,c):
    if a>b and a>c:
        return a
    elif b>a and b>c:
        return b
    elif c>a or c>b:
        return c
    else:
        return a or b or c
def main():
    a = int(input('Enter 1st side: '))
    b = int(input('Enter 2nd side: '))
    c = int(input('Enter 3rd side: '))
    d = find_max(a,b,c)
    print(d)
main()   
```

```python
# Write a function, is_vowel that returns the value true if a given character is a vowel, 
# and otherwise returns false.
def is_vowel(c):
    if c in ['A', 'a', 'E', 'e', 'I', 'i', 'O', 'o', 'U', 'u']:
        return True
    else:
        return False
# Write another function main, in main() function accept a string from user and count number 
# of vowels in that string.

def main():
    vowel = 0
    abc = input()
    for character in abc:
      if is_vowel(character):
         vowel+=1
    return vowel
main()
```

```python
# For a given list of values in descending order, write a method in python to search 
# for a value with the help of Binary Search method. 
# The method should return position of the value and should return -1 if the value not present in the list.

#Using Recursion
def search_element_recursion(lst, l, r, value_to_search):
    middle_index = l + (r - l)//2
    if r >= l:
      if value_to_search == lst[middle_index]:
        return middle_index
      if value_to_search < lst[middle_index]:
        return search_element_recursion(lst, middle_index + 1, r, value_to_search)
      else:
          return search_element_recursion(lst, l, middle_index - 1, value_to_search)
    else:
        return -1

#Using Iterative Search
def search_element_iterative(lst, l, r, value_to_search):
    
    while r >= l:

        middle_index = l + (r - l)//2
        if value_to_search == lst[middle_index]:
            return middle_index
        if value_to_search < lst[middle_index]:
            l = middle_index + 1
        else:
             r = middle_index - 1
   
    return -1  

lst=[int(i) for i in input().strip().split(',')]
value_to_search = int(input('Enter element to search: '))
#Using Recursion
print(f'Using Recursion Method: {search_element_recursion(lst, 0, len(lst)-1, value_to_search)}')
#Using Iterative
print(f'Using Iterative Method: {search_element_iterative(lst, 0, len(lst)-1, value_to_search)}')
```

```python
# Write a function half_and_half that takes in a list and change the list 
# such that the elements of the second half are now in the first half.

# For example, if the size of list is even and content of list is as follows :
# my_liist = [10,20,30,40,50,60]
# The output should be
# [40,50,60,10,20,30]
# if the size of list is odd and content of list is as follows :
# my_liist = [10,20,30,40,50,60,70]
# The output should be
# [50,60,70,40,10,20,30]


def half_and_half(lst):
    middle_index = int(len(lst) /2)
    return lst[middle_index : len(lst)] + lst[: middle_index] if len(lst) % 2 == 0 else lst[middle_index+1 : len(lst)] + lst[middle_index:middle_index + 1] + lst[: middle_index]

lst = [int(i) for i in input().strip().split(',')]
print(half_and_half(lst))
```

```python
# Write a function that accepts a dictionary as an argument. If the dictionary contains replicate values, 
# return an empty dictionary, otherwise, return a new dictionary whose values are now the keys and whose 
# keys are the values.
def check_dict(dict):
    a = set()
    new_dict = {}
    for key in dict:
        if dict[key] not in a:
           a.add(dict[key])
        else:
            return {} 
        new_dict.update({dict[key] : key})
    return new_dict

keys = [i for i in input('Enter keys of the dictionary:').split(',')]
values = [i for i in input('Enter values of the dictionary:').split(',')]
print(check_dict({k : v for k, v in zip(keys,values)}))
# Enter keys of the dictionary:1,2,3,100,0
# Enter values of the dictionary:Apple,Apple,Mangoes,Orange,Name    
        
```
